```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import itertools
from pylab import *
```

```python
lengths = linspace(0.5,3.5,500)
energies = 2*(-1*np.exp(-np.power(lengths/2,2))) + 1*1*np.exp(-np.power(lengths,2))
e_energies = 2*(-1*np.exp(-np.power(lengths/2,2)))
nuc_energies = 1*1*np.exp(-np.power(lengths,2))

# Increasing the 1's to 8's increases the bond distance.  The minimum here looks to be about 1.
# That is about right given the pseudopotentials here.  Better pseudopotentials would increase
# the accuracy.

plt.plot(lengths,energies)
plt.plot(lengths,e_energies,'g')
plt.plot(lengths,nuc_energies,'r')
```

```python
O2_energies = 2*(-8*np.exp(-np.power(lengths/2,2))) + 8*8*np.exp(-np.power(lengths,2))
O_2_e_energies = 2*(-8*np.exp(-np.power(lengths/2,2)))
O_2_nuc_energies = 8*8*np.exp(-np.power(lengths,2))
```

```python
plt.plot(lengths*0.52917721067121,energies*10,'k',label=r'$H_2\times10$') #Conversion factor is atomic units of length to Angstroms
plt.plot(lengths*0.52917721067121,O2_energies,'r',label=r'$O_2$')
plt.plot(lengths*0.52917721067121,np.zeros(len(lengths)),'k--')
plt.xlabel(r'bond length ($\AA$)')
plt.ylabel('energy atomic units (Hartree)')
plt.legend()
plt.savefig("diatomic_molecules.png",dpi=220)
```

```python
plt.plot(lengths,energies,'k',label=r'$H_2$')
plt.xlabel('bond length atomic units')
plt.ylabel('energy atomic units (Hartree)')
plt.legend()
print(lengths[np.argmin(energies)]*0.52917721067121)
```
